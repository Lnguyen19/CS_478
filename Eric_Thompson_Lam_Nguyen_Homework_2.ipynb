```python
#
# HW2 - Creating Models
# Lam Nguyen and Eric Thompson
#
##########
# Part 1 #
##########
# Dataset description
# The inputs are 50,000 color training images, and 10,000 test images
# The labels represent a description for what the image is of, IE airplane/automobile/etc.
# It is a 50,000 image dataset of 32x32 pixels, in 3 channels (RGB)
#   Dimensions are 50,000 x 32 x 32 x 3 for training
#   10,000 x 32 x 32 x 3 for testing

import keras
from keras.datasets import cifar10
from keras import activations
from keras.models import Sequential
from keras import layers
from keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from keras.layers import Conv2D



##########
# Part 2 #
##########
# Split into train/test
(xTrain, yTrain), (xTest, yTest) = keras.datasets.cifar10.load_data()
#Reshaping to correct dimensions
assert xTrain.shape == (50000, 32, 32, 3)
assert xTest.shape == (10000, 32, 32, 3)
assert yTrain.shape == (50000, 1)
assert yTest.shape == (10000,1)
#Convert datasets to float value between 0 and 1
xTrain = xTrain.reshape(50000, 3072).astype('float32')
xTrain /= 255
xTest = xTest.reshape(10000, 3072).astype('float32')
xTest /= 255
yTrain = to_categorical(yTrain, num_classes=10)
yTest = to_categorical(yTest, num_classes=10)



##########
# Part 3 #
##########

# ReLu
modelReLu = Sequential()
modelReLu.add(Dense(64, input_dim=3072, activation = 'relu'))
modelReLu.add(Dense(10, activation = 'relu'))
modelReLu.summary()

# TanH - 0 centered output helps network rapidly learn
modelTanH = Sequential()
modelTanH.add(Dense(64, input_dim=3072, activation = 'tanh'))
modelTanH.add(Dense(10, activation = 'tanh'))
modelTanH.summary()

# ELU - Chosen for its ability to learn quickly
modelELU = Sequential()
modelELU.add(Dense(64, input_dim=3072, activation = 'elu'))
modelELU.add(Dense(10, activation = 'elu'))
modelELU.summary()



##########
# Part 4 #
##########

# Using categorical_crossentropy loss function with adam optimization

# ReLu
modelReLu.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
modelReLu.fit(xTrain, yTrain, epochs=20, batch_size=100)
accuracy = modelReLu.evaluate(xTest , yTest)
print("ReLu Accuracy: " + str(accuracy))
# Accuracy: 24.59%

# TanH
modelTanH.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
modelTanH.fit(xTrain, yTrain, epochs=20, batch_size=100)
accuracy = modelTanH.evaluate(xTest , yTest)
print("TanH Accuracy: " + str(accuracy))
# Accuracy: 10.00%

# ELU
modelELU.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
modelELU.fit(xTrain, yTrain, epochs=20, batch_size=100)
accuracy = modelELU.evaluate(xTest , yTest)
print("ELU Accuracy: " + str(accuracy))
# Accuracy: 19.97%



##########
# Part 5 #
##########

# Remaking the ReLu

# Creating the model
model = Sequential()
#modelTanH.add(Dense(128, input_dim=3072, activation = 'relu')) 
#modelTanH.add(Dense(256, input_dim=3072, activation = 'relu'))
model.add(Dense(512, input_dim=3072, activation = 'relu'))
model.add(Dense(64, input_dim=3072, activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.summary()

#Compile/fit/evaluate
model.compile(loss = 'kl_divergence', optimizer = 'adadelta', metrics = ['accuracy'])
model.fit(xTrain, yTrain, epochs=20, batch_size=100)
accuracy = model.evaluate(xTest , yTest)
print("New Model ReLu Accuracy: " + str(accuracy))
# Accuracy: 10.01%

# Conclusion: The changes, overall, did not improve the accuracy of the model. categorical_crossentropy seems to be ideal for dealing with multiple
# different classes.
```
