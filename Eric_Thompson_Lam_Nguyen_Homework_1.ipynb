Activation Function : Sigmoid

g(x) = sigmoid.JPG

```python
import numpy as np
import matplotlib.pyplot as plt
from math import e #Euler's number

#Initialize arrays
x = np.arange(-5, 5, 0.01)
y = []

#Returns sigmoid equation with exponentValue as the value for x
def SigmoidEquation(exponentValue):
  calcVal = 1 + e**-exponentValue
  return (1 / calcVal)


#Loop through every value in x, set y = 1/(1+e^-x)
for num in x:
  y.append(SigmoidEquation(num))

#Create graph
plt.plot(x,y) 
plt.title('Sigmoid')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function : Tahn

g(x) = tahn.JPG

```python
import numpy as np
import matplotlib.pyplot as plt
from math import e #Euler's number

#Initialize arrays
x = np.arange(-5, 5, 0.01)
y = []

#Returns tahn equation with exponentValue as value for x
def TahnEquation(exponentValue):
  calcValTop = (e**exponentValue) - (e**-exponentValue)
  calcValBottom = (e**exponentValue) + (e**-exponentValue)
  return calcValTop/calcValBottom


#Loop through every value in x, set y = 1/(1+e^-x)
for num in x:
  y.append(TahnEquation(num))

#Create graph
plt.plot(x,y) 
plt.title('Tahn')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function : Gelu

g(x) = GELU.JPG

```python
import numpy as np
import matplotlib.pyplot as plt
import math #For math.erf and math.sqrt


#Initialize arrays
x = np.arange(-5, 5, 0.01)
y = []

#Runs through Gelu equation with val for x
def GeluEq(val):
  tempVal = 1 + math.erf(val/math.sqrt(2))
  return (1/2)*val*tempVal

#Loop through every value in x, if x < 0, y = alpha * x, else y = x
for num in x:
  y.append(GeluEq(num))

#Create graph
plt.plot(x,y) 
plt.title('Gelu ')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function : Softplus

g(x) = softplus.JPG

```python
# @TODO Softplus activation function
import numpy as np
import matplotlib.pyplot as plt
from math import e #Euler's number
x = np.arange(-5, 5, 0.01)
y = []

def exponeting(value):
  value = e**value
  return value


for num in x:
  num = np.log(1+exponeting(num))
  y.append(num)
  


plt.plot(x,y) 
plt.title('Softplus')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()


```

Activation Function : Elu

let alpha = 1

g(x) = elu.JPG

```python
# @TODO ELU activation function
import numpy as np
import matplotlib.pyplot as plt
from math import e #Euler's number
a = 1
x = np.arange(-5, 5, 0.01)
y = []
def exponeting(value):
  value = e**value
  return value

for num in x:
  if num <=0:
    num = a*(exponeting(num)-1)
  y.append(num)

plt.plot(x,y) 
plt.title('elu')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()


```

```python
# @TODO SELU activation function
import numpy as np
import matplotlib.pyplot as plt
from math import e #Euler's number
x = np.arange(-5, 5, 0.01)
y = []
a = 1.67326
lamda = 1.0507
def exponeting(value):
  value = e**value
  return value

for num in x:
  if(num<0):
    num = lamda*(a* (exponeting(num)-1) )
  y.append(num)
  
plt.plot(x,y) 
plt.title('selu')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()

    
```

Activation Function : SiLU

g(x) = SILU.JPG

```python
# @ TODO SiLU Activation function
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.01)
y = []

def exponeting(value):
  value = e**(0-value)
  return value

for num in x:
  num = num/(1+exponeting(num))
  y.append(num)
plt.plot(x,y) 
plt.title('SiLU')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function : Gaussian

```python
import numpy as np
import matplotlib.pyplot as plt
from math import e #Euler's number

x = np.arange(-5, 5, 0.01)
y = []

for num in x:
  y.append(e**(-(num**2)))

plt.plot(x,y) 
plt.title('Gaussian')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```
